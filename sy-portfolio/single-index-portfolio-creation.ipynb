## Steps to do a single index model

- __Compute the inital position of each security in the active portfolio:__

``` security weight = security alpha / residual variance of security```
- __Scale those initial position to force portfolio weights to sum to 1 by dividing by their sum, that is:__

``` portfolio weight security = security weight / sum of all weights```

- __Compute the alpha of the active portfolio:__

``` active portfolio alpha = weighted average securities alphas```

- __Compute the residual variance of the active portfolio:__

``` active portfolio residual variance = weighted average residual variance of the active securities```

- __Active portfolio weight of the Active + Passive portfolio:__

``` portfolio weight active = (portfolio alpha / active portfolio residual variance) / (expected market premium / market variance)```

- __Compute the beta of the Active portfolio__

``` active beta = weighted average of the securities beta ```

- __Adjust the initial position of the active portfolio__

``` active weight = portfolio weight active / (1 + (1- active beta) * portfolio weight active)
    passive weight = 1- active weight ```
